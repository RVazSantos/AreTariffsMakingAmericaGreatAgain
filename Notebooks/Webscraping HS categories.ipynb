```python
import pandas as pd

import requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36"
}

url = "https://exporterai.com/chapters.html"
response = requests.get(url, headers=headers)
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, "html.parser")
```

```python
section = soup.find('section', id='chapter-table')
table = section.find('table', class_='table custom-table')
table
```

```python
rows = table.find_all('tr')
rows
```

```python
heads = table.find_all("th")
columns = [th.get_text(strip=True) for th in heads]

chapters = pd.DataFrame(columns=columns)
chapters
```

```python
data = [
    [cell.get_text(strip=True) for cell in row.find_all(["td", "th"])]
    for row in table.find_all("tr")
]

chapters = pd.DataFrame(data[1:], columns=data[0])
chapters
```

```python
chapters["InitRange"] = chapters["Chapter Range"].str[0:2]
chapters
```

```python
chapters["FinalRange"] = chapters["Chapter Range"].str[-2:]
chapters
```
