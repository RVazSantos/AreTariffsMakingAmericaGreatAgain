```python
import pandas as pd
```

```python
currency = pd.read_excel(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\Indexs\CurrencyTable.xlsx")
currency.head(5)
```

```python
currency.rename(columns={"EXCAUS":"CAD_USD",
                         "EXCHUS":"CNY_USD",
                         "EXMXUS":"MXN_USD",
                         "EXUSEU":"EUR_USD"},inplace=True)
```

```python
currency["Month"] = currency["observation_date"].dt.month
```

```python
currency["Year"] = currency["observation_date"].dt.year
```

```python
def q(row):
    if row["Month"] in (1,2,3):
        return "Q1"
    elif row["Month"] in (4,5,6):
        return "Q2"
    elif row["Month"] in (7,8,9):
        return "Q3"
    elif row["Month"] in (10,11,12):
        return "Q4"
    else:
        return "Error!"
```

```python
currency["Quarter"] = currency.apply(q,axis=1)
currency.head(5)
```

```python
CAD_USD_table = currency.pivot_table(index=["Year","Quarter"],values="CAD_USD",aggfunc="mean").reset_index()
CAD_USD_table.rename(columns={"CAD_USD":"Value"},inplace=True)
CAD_USD_table["Country"] = "Canada"
CAD_USD_table["Index"] = "Currency"
CAD_USD_table["Unit"] = "CAD_USD"
CAD_USD_table
```

```python
CNY_USD_table = currency.pivot_table(index=["Year","Quarter"],values="CNY_USD",aggfunc="mean").reset_index()
CNY_USD_table.rename(columns={"CNY_USD":"Value"},inplace=True)
CNY_USD_table["Country"] = "China"
CNY_USD_table["Index"] = "Currency"
CNY_USD_table["Unit"] = "CNY_USD"
CNY_USD_table
```

```python
MXN_USD_table = currency.pivot_table(index=["Year","Quarter"],values="MXN_USD",aggfunc="mean").reset_index()
MXN_USD_table.rename(columns={"MXN_USD":"Value"},inplace=True)
MXN_USD_table["Country"] = "Mexico"
MXN_USD_table["Index"] = "Currency"
MXN_USD_table["Unit"] = "MXN_USD"
MXN_USD_table
```

```python
EUR_USD_table = currency.pivot_table(index=["Year","Quarter"],values="EUR_USD",aggfunc="mean").reset_index()
EUR_USD_table.rename(columns={"EUR_USD":"Value"},inplace=True)
EUR_USD_table["Country"] = "European Union"
EUR_USD_table["Index"] = "Currency"
EUR_USD_table["Unit"] = "EUR_USD"
EUR_USD_table
```

```python
index_table = pd.read_excel(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\Indexs\index table.xlsx")
index_table = index_table[["Year","Quarter","Country","Index","Value","Unit"]]
index_table.head(5)
```

```python
index_table2 = pd.concat([index_table, CAD_USD_table, CNY_USD_table, MXN_USD_table, EUR_USD_table])
index_table2
```
