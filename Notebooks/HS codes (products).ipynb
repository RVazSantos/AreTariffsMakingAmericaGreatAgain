```python
### This notebook has the purpose of loading several pdf files of US Tariff Proclamations in order to retrieve the HS codes affected in each 
```

```python
import pandas as pd
import pdfplumber
import re

pdf_path = r"C:\Users\rcrds\Desktop\Ironhack\Final Project\steelHTSlist final (1).pdf"

steel_codes = set()

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        
        if text:
            matches = re.findall(r"\b(72\d{2}|73\d{2}|84\d{2}|85\d{2}|94\d{2})\b", text)

            for m in matches:
                steel_codes.add(m)
steel_hs4 = sorted(steel_codes)

print("Steel HS4 Codes Covered:")
print(steel_hs4)
```

```python
pdf_path2 = r"C:\Users\rcrds\Desktop\Ironhack\Final Project\aluminumHTSlist final.pdf"

aluminum_codes = set()

with pdfplumber.open(pdf_path2) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        
        if text:
            matches = re.findall(r"\b(76\d{2}|87\d{2}|66\d{2}|83\d{2}|84\d{2}|85\d{2}|88\d{2}|90\d{2}|94\d{2}|95\d{2}|96\d{2})\b", text)

            for m in matches:
                aluminum_codes.add(m)
aluminum_hs4 = sorted(aluminum_codes)

print("Aluminum HS4 Codes Covered:")
print(aluminum_hs4)
```

```python
pdf_path3 = r"C:\Users\rcrds\Desktop\Ironhack\Final Project\autoparts final.pdf"

auto_parts_codes = set()

with pdfplumber.open(pdf_path3) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        
        if text:
            matches = re.findall(r"\b(87\d{2}|40\d{2}|73\d{2}|83\d{2}|84\d{2}|85\d{2}|70\d{2}|90\d{2}|94\d{2})\b", text)

            for m in matches:
                auto_parts_codes.add(m)
auto_parts_hs4 = sorted(auto_parts_codes)

print("Auto Parts HS4 Codes Covered:")
print(auto_parts_hs4)
```

```python
automotive_codes = ["8703","8704"]
automotive_codes
```

```python
steel_df = pd.DataFrame(steel_hs4)
steel_df.rename(columns={0:"HS4"},inplace=True)
steel_df["Tariff"] = "Applied"
steel_df["Scope"] = "Steel"
steel_df
```

```python
aluminum_df = pd.DataFrame(aluminum_hs4)
aluminum_df.rename(columns={0:"HS4"},inplace=True)
aluminum_df["Tariff"] = "Applied"
aluminum_df["Scope"] = "Aluminum"
aluminum_df
```

```python
autoparts_df = pd.DataFrame(auto_parts_hs4)
autoparts_df.rename(columns={0:"HS4"},inplace=True)
autoparts_df["Tariff"] = "Applied"
autoparts_df["Scope"] = "Auto Parts"
autoparts_df
```

```python
automotive_df = pd.DataFrame(automotive_codes)
automotive_df.rename(columns={0:"HS4"},inplace=True)
automotive_df["Tariff"] = "Applied"
automotive_df["Scope"] = "Automotive"
automotive_df
```

```python
AffectedHS = pd.concat([steel_df,aluminum_df,autoparts_df,automotive_df]).sort_values(by="HS4")
AffectedHS.tail(50)
```
