```python
def aloc_section(df1,df2):
    result = []

    for hs in df1["HS2"].astype(int):
        match = df2[(df2["InitRange"].astype(int) <= hs) &(df2["FinalRange"].astype(int) >= hs)]
        if len(match) > 0:
            result.append(match.iloc[0]["Section No."])
        else:
            result.append(None)
    return result
        
HS_codes["Section"] = aloc_section(HS_codes,chap)
HS_codes
```
