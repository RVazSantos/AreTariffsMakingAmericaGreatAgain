```python
import pandas as pd

df_exports = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\2020-2024 months with Com\Exports_2020_2024_HS2.csv" , encoding="latin1")
df_exports
```

```python
col_drop2 = ["cmdDesc","isLeaf","qty","classificationSearchCode","altQty","partner2Desc","classificationCode","aggrLevel","mosCode","customsCode","customsDesc","partner2Code","partner2ISO","motCode","motDesc","qtyUnitCode","typeCode","period","reporterISO","reporterDesc","netWgt","isNetWgtEstimated","grossWgt","isGrossWgtEstimated","legacyEstimationFlag","isReported","isAggregate","qtyUnitAbbr","flowDesc","partnerDesc","isQtyEstimated","altQtyUnitCode","altQtyUnitAbbr","isAltQtyEstimated"]
df_exports.drop(columns=col_drop2, inplace=True)
df_exports
```

```python
df_exports = df_exports[df_exports["partnerISO"]!="World"]
```

```python
df_exports = df_exports.rename(columns = {'freqCode':'YearMonthDay',
                                    "refPeriodId":"Year",
                                    "refYear":"Month",
                                    "refMonth":"YearMonth",
                                    "isOriginalClassification":"HS2",
                                    "cmdCode":"HS2 Description"})
```

```python
export_pivot = df_exports.pivot_table(index="partnerISO", columns="Year", values="fobvalue", aggfunc="sum")
```

```python
export_pivot = export_pivot / 1_000_000  # convert to millions
export_pivot = export_pivot.round(2)
export_pivot.fillna(0,inplace=True)
export_pivot
```

```python
df_imports = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Clean Datasets\Imports_2020_2025.csv",dtype={"HS4": str}).drop(columns=["Unnamed: 0","cmdDesc","primaryValue"])
df_imports = df_imports[df_imports["partnerISO"]!="World"]
df_imports2 = df_imports
df_imports = df_imports[df_imports["Year"]!=2025]
df_imports
```

```python
import_pivot = df_imports.pivot_table(index="partnerISO", columns="Year", values="fobvalue", aggfunc="sum")
import_pivot = import_pivot / 1_000_000  # convert to millions
import_pivot = import_pivot.round(2)
import_pivot.fillna(0,inplace=True)
import_pivot
```

```python
export_pivot
```

```python
diff = export_pivot.subtract(import_pivot).reset_index()
diff.fillna(0,inplace=True)
diff.head(50)
```
