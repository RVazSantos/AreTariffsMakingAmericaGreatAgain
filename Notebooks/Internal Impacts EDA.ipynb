```python
import pandas as pd 
```

```python
TradeData_2023 = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\TradeData_2023.csv")
TradeData_2023 = TradeData_2023[["refPeriodId","refYear","refMonth","flowCode","fobvalue"]]
TradeData_2023["fobvalue"] = TradeData_2023["fobvalue"]/1000000
TradeData_2024 = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\TradeData_2024.csv")
TradeData_2024 = TradeData_2024[["refPeriodId","refYear","refMonth","flowCode","fobvalue"]]
TradeData_2024["fobvalue"] = TradeData_2024["fobvalue"]/1000000
TradeData_2025 = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Final Project\Original Datasets\TradeData_2025.csv")
TradeData_2025 = TradeData_2025[["refPeriodId","refYear","refMonth","flowCode","fobvalue"]]
TradeData_2025["fobvalue"] = TradeData_2025["fobvalue"]/1000000
TradeData = pd.concat([TradeData_2023,TradeData_2024,TradeData_2025])
TradeData
```

```python
TradeData2 = TradeData.pivot_table(index=["refPeriodId","refYear","refMonth"], columns="flowCode", values="fobvalue", aggfunc="sum").reset_index()
TradeData2["Balance"] = TradeData2["Export"] - TradeData2["Import"]
TradeData2.head()
```
